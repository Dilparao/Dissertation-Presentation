```python
def transform_age(age):
    if age >= 18 and age <= 37:
        return 1
    elif age >= 38 and age <= 45:
        return 2
    elif age >= 46 and age <= 53:
        return 3
    elif age >= 54 and age <= 61:
        return 4
    elif age >= 62 and age <= 69:
        return 5
    elif age >= 70 and age <= 80:
        return 6

clean_df = clean_df.assign(final_age = clean_df['age'].apply(transform_age))
clean_df.tail()
    
```

```python
def transform_hp(ap_hi):
    if ap_hi >= 60 and ap_hi <= 89:
        return 1
    elif ap_hi >= 90 and ap_hi <= 119:
        return 2
    elif ap_hi >= 120 and ap_hi <= 129:
        return 3
    elif ap_hi >= 130 and ap_hi <= 139:
        return 4
    elif ap_hi >= 140 and ap_hi <= 179:
        return 5
    elif ap_hi >= 180 and ap_hi <= 370:
        return 6

clean_df = clean_df.assign(final_ap_hi = clean_df['ap_hi'].apply(transform_hp))
clean_df.tail()
    
```

```python
def transform_lp(ap_lo):
    if ap_lo >= 30 and ap_lo <= 60:
        return 1
    elif ap_lo >= 61 and ap_lo <= 79:
        return 2
    elif ap_lo >= 80 and ap_lo <= 89:
        return 3
    elif ap_lo >= 90 and ap_lo <= 99:
        return 4
    elif ap_lo >= 100 and ap_lo <= 109:
        return 5
    elif ap_lo >= 110 and ap_lo <= 360:
        return 6

clean_df = clean_df.assign(final_ap_lo = clean_df['ap_lo'].apply(transform_lp))
clean_df.tail()
    
```

```python
def transform_bmi(bmi):
    if bmi <= 18 :
        return 1
    elif bmi >= 19 and bmi <= 24:
        return 2
    elif bmi >= 25 and bmi <= 29:
        return 3
    elif bmi >= 30 and bmi <= 100:
        return 4
   

clean_df = clean_df.assign(final_bmi = clean_df['bmi'].apply(transform_bmi))
clean_df.tail()
    
```

```python
Y = clean_df.iloc[:,11].values
X = clean_df.drop(['cardio', 'height','smoke','active','alco','gender','pulse','age','weight','ap_hi','ap_lo','bmi'], axis =1)
```

```python
#req_model='SvmRbf'
def get_predictions(cholesterol, gluc,age,  ap_hi, ap_lo, bmi, req_model):
    mylist = [cholesterol, gluc, age, ap_hi, ap_lo, bmi]
    mylist = [float(i) for i in mylist]
    vals = [mylist]

    if req_model == 'Logistic':
        #print(req_model)
        return logistic_model.predict(vals)[0]

    elif req_model == 'DecisionTree':
         #print(req_model)
        return dt_model.predict(vals)[0]

    elif req_model == 'SvmLinear':
         #print(req_model)
        return svm_linear_model.predict(vals)[0]
    
    elif req_model == 'SvmPoly':
         #print(req_model)
        return svm_poly_model.predict(vals)[0]
    
    elif req_model == 'SvmRbf':
        #print(req_model)
        return svm_rbf_model.predict(vals)[0]    
    
    elif req_model == 'RandomForest':
          #print(req_model)
        return rf_model.predict(vals)[0]    
    else:
        return "Cannot Predict"
```
